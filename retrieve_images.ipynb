```python
import wget
import os

def retrieve_image(image_url, folder):
    filename = os.path.join(folder, image_url.split("/")[-1])
    try:
        image_filename = wget.download(image_url, out=filename)
        print('Image sucessfully Downloaded: ',filename)
    except:
        print('Image Couldn\'t be retreived')
```

```python
with open('imagenet.synset.txt', 'r', encoding='utf-8') as f:
    image_urls = f.read().split('\n')[:-1]
```

```python
for image_url in image_urls:
    retrieve_image(image_url, 'data')
```
